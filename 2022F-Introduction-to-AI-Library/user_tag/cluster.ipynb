```python
data["book_cnt"] = np.log(data["book_cnt"] + 1)

data["in_library_time"] = np.log(data["in_library_time"] + 1)

data["room_time"] = np.log(data["room_time"] + 1)

data["in_library_day"] = np.log(data["in_library_day"] + 1)

data["room_day"] = np.log(data["room_day"] + 1)
```

```python
# X = data.drop(["sid"], axis=1)
# X = data[["book_cnt"]]
X = data[["in_library_time", "room_time"]]

X.head()
```

```python
num = 3
```

### KMeans

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=num, random_state=0)
km.fit(X)

predict = km.predict(X)

cnts = np.zeros(num)
for i in range(num):
    cnts[i] = np.where(predict == i)[0].shape[0]

cnts 
```

```python
centers = km.cluster_centers_

display(centers)
```

```python
max_distance = np.zeros((num, X.shape[1]))

for k in range(num):
    cluster_k = X[predict == k]
    max_distance[k] = np.max(np.sqrt((cluster_k - centers[k]) ** 2))

max_distance
```
